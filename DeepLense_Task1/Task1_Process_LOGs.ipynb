```python
from task1Utils import load_dict, plot_loss
import os
```

```python
name_list = [
    'task1-EfNetB2-RGB-150-LR-0.001-gamma-0.2-step-5-epochs-30',
    'task1-ResNet18-RGB-150-LR-0.001-gamma-0.2-step-5-epochs-30',
]
```

```python
losses_list = []

for name in name_list:
    checkpoint_dir = './checkpoints_'+name
    model_name = name.split('-')[1]
    log_path = os.path.join(checkpoint_dir, name+'-LOG.pkl')
    log = load_dict(log_path)
    losses_list.append(log['test_losses'])
    print(model_name, log['best_accuracy'])
```

```python
plot_loss(name_list, losses_list, figw=15, figh=7.5, linewidth=2, fontsize=15)
```
